```python
from ultralytics import YOLO
model = YOLO("yolov8n.yaml")
```

```python
import yaml

yaml_dict = {
    "train": trainPath,
    "val": valPath,
    "nc": 9,
    "names": ['SH','SP','SC','OP','MB','HB','CS','CFO','BMFO']
}

# Define the path for the YAML file
yaml_path = "data.yaml"

# Create the YAML file
with open(yaml_path, 'w') as file:
    yaml.dump(yaml_dict, file)
```

```python
project_path = "/content/drive/MyDrive/pcb_train_results"
name_project = "train"

results = model.train(
    data=yaml_path,
    epochs=100,
    imgsz=640,
    batch=96,
    pretrained=False,
    project=project_path,
    name=name_project
)
```
